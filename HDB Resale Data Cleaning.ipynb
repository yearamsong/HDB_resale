```python
#Import all csv files

datapath = "C://Users//YR//Desktop//DataProjects//HDB_resale//Data//"

import glob

df = pd.concat([pd.read_csv(f) for f in glob.glob(datapath+'*.csv')], ignore_index = True)
df.head()
```

```python
#replace flat_type MULTI-GENERATION TO MULTI GENERATION
df['flat_type'] = df['flat_type'].replace(['MULTI-GENERATION'],'MULTI GENERATION')
print(df['flat_type'].unique())
```

```python
#change to uppercase
df['flat_model']=df['flat_model'].str.title()
print(df['flat_model'].unique())
```

```python
#rename month to Sales yyyy-mm
df = df.rename({'month': 'sales_yyyy-mm'}, axis=1)

#add sales_yyyy, sales_mm and remaining_lease_calc
df['sales_yyyy']=df['sales_yyyy-mm'].str[:4]
df['sales_mm']=df['sales_yyyy-mm'].str[-2:]
df['remaining_lease_calc']=pd.to_numeric(df['sales_yyyy'])-pd.to_numeric(df['lease_commence_date'])
df['address query']=df['block']+" "+df["street_name"]
df.head()
```

```python
###Sample Script to extract longitude/latitude information given street name
import json
import requests
import time
from tqdm import tqdm_notebook
import openpyxl

count = 0
loop_round = 0

#open workbook
wb = openpyxl.load_workbook(datapath+'export_query_output_vF.xlsx')
sheet = wb['Sheet1']

for row in tqdm_notebook(range(2813, sheet.max_row + 1)):
        query_address = sheet.cell(row,1).value
        query_string = "https://developers.onemap.sg/commonapi/search?searchVal="+str(query_address)+"&returnGeom=Y&getAddrDetails=Y&pageNum=1"
        result = requests.get(query_string)

        # Convert JSON into Python Object
        data=json.loads(result.content)

        # Extract data from JSON
        try:
            sheet['B' + str(row)]=data['results'][0]['LONGITUDE']
            sheet['C' + str(row)]=data['results'][0]['LATITUDE']
            sheet['D' + str(row)]=data['results'][0]['SEARCHVAL']
            sheet['E' + str(row)]=data['results'][0]['BLK_NO']
            sheet['F' + str(row)]=data['results'][0]['ROAD_NAME']
            sheet['G' + str(row)]=data['results'][0]['BUILDING']
            sheet['H' + str(row)]=data['results'][0]['ADDRESS']
            sheet['I' + str(row)]=data['results'][0]['POSTAL']
            sheet['J' + str(row)]= 1
            print(str(loop_round)+"-"+str(count)+": "+ str(query_address) + data['results'][0]['LATITUDE'] + data['results'][0]['LONGITUDE'])

        except:
            sheet['D' + str(row)]="Error"
            print(str(loop_round)+"-"+str(count)+": "+ str(query_address) + "NOT FOUND")
            pass

        count = count + 1
        
        #rest 15 seconds every 100 iteration
        if count ==100:
            print("Pausing for 15 seconds")
            wb.save(datapath+"export_query_output_vF.xlsx")
            time.sleep(15)
            count=0
            loop_round +=1
        else:
            continue
            
        #rest 60 seconds every 5 loops
        #if loop_round%5 == 0 and loop_round >0:
        #    print("Pausing for 60 seconds")
        #    wb.save(datapath+"export_query_output_vF.xlsx")
        #    time.sleep(60)
        #    loop_round +=1
        #else:
        #    continue

print('Job Done')
        
```
